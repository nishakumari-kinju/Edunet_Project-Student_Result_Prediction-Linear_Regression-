```python
%pip install scikit-learn
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor 
```

```python
features = ['study_hours','social_media_hours',
'gaming_hours', 'screen_time_hours', 'part_time_job',
'focus_index', 'burnout_level',
'productivity_score']
```

```python
target = "exam_score"
```

```python
df_model = df[features + [target]].copy()
```

```python
df_model
```

```python
le = LabelEncoder()
```

```python
df_model["part_time_job"] = le.fit_transform(df_model["part_time_job"])
df_model
```

```python
X = df_model[features]
```

```python
Y = df_model[target]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
len(Y_train)
```

```python
len(Y_test)
```

```python
model = {
    "LinearRegression": {
        "model": LinearRegression(),
        "params": {}
    },
    "DecisionTree" : {
        "model" : DecisionTreeRegressor(),
        "params": {"max_depth": [3,5,10], "maint_sample_split": [2,5]}
        },
    "RandomForest" : {
        "model" : RandomForestRegressor(),
        "params": {"n_estimators": [50,100], "max_depth": [5,10]}
    }
}
```

```python
best_models = []
```

```python
model = {
    "LinearRegression": {
        "model": LinearRegression(),
        "params": {}
    },
    "DecisionTree" : {
        "model" : DecisionTreeRegressor(),
        "params": {"max_depth": [3,5,10], "min_samples_split": [2,5]} # Corrected parameter name here
        },
    "RandomForest" : {
        "model" : RandomForestRegressor(),
        "params": {"n_estimators": [50,100], "max_depth": [5,10]}
    }
}

for name, config in model.items():
  print(f"Training {name}")

  grid = GridSearchCV(config["model"], config["params"], cv=5, scoring="neg_mean_squared_error")
  grid.fit(X_train, Y_train)

  Y_pred = grid.predict(X_test)
  rmse= np.sqrt(mean_squared_error(Y_test, Y_pred))
  r2= r2_score(Y_test, Y_pred)

  best_models.append({
      "model":name,
      "best_params": grid.best_params_,
      "rmse": rmse,
      "r2":r2
  })
```

```python
best_models
```

```python
result_df = pd.DataFrame(best_models)
```

```python
result_df.sort_values(by="rmse")
```

```python
import joblib

best_row = result_df.sort_values(by="rmse").iloc[0]
```

```python
best_row
```

```python
best_model_name = best_row["model"]
```

```python
best_model_config = model[best_model_name]
```
